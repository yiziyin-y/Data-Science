- Double Exponential Smoothing

```python
def double_exponential_smoothing(series, alpha, beta):
    result = [series[0]]
    
    for n in range(1, len(series)+1):
        if n == 1:
            level, trend = series[0], series[1] - series[0]
        if n >= len(series): # forecasting
            value = result[-1]
        else:
            value = series[n]
        
        last_level, level = level, alpha * value + (1 - alpha) * (level + trend)
        trend = beta * (level - last_level) + (1 - beta) * trend
        result.append(level + trend)
    return result

def plot_double_exponential_smoothing(series, alphas, betas):
    plt.figure(figsize=(17, 8))
    for alpha in alphas:
        for beta in betas:
            plt.plot(double_exponential_smoothing(series, alpha, beta), label="Alpha {}, beta {}".format(alpha, beta))
    plt.plot(series.values, label = "Actual")
    plt.legend(loc="best")
    plt.axis('tight')
    plt.title("Double Exponential Smoothing")
    plt.grid(True)

plot_double_exponential_smoothing(BILI.Close, alphas=[0.9, 0.04], betas=[0.9, 0.04])
```

```python
### Data transformations
BILI.columns = [x.strip() for x in BILI.columns]
BILI.columns = [x.replace(' ', '_') for x in BILI.columns]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
BILI.columns
```

```python
pd.concat([BILI.Close, BILI.Close.shift(-1)], axis=1)
```

```python
def valid(col):
    return col
BILI = BILI.dropna()
X = BILI[[x for x in BILI.columns if valid(x)]].iloc[1:]
y = BILI.Close.shift(-1).dropna()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X.T
```

```python
rf_model = RandomForestRegressor()
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_model.score(X_test, y_test)
```
